Import training data, and libraries for several classifiers

````python
# read the trainings data
dtrain = pd.read_csv('./data/flutype_train.csv', sep="\t")
# transform data
dtrain_1=pd.get_dummies(dtrain)
X_train = pd.DataFrame(dtrain_1, columns = ['P1','P3','P4','P5','P6','P7','P8','P10','P13'])
Y_train = pd.DataFrame(dtrain_1, columns = ['virus_H1','virus_H5','virus_H7','virus_NY','virus_Vi','virus_X31'])
````

````python
#import libraries for different classifiers
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
% matplotlib inline
import time 
from sklearn import decomposition

# build several classifiers with trainings dataset

classifiers = [
    KNeighborsClassifier(3), #three nearest neighbors
    DecisionTreeClassifier(random_state=1),
    RandomForestClassifier(n_estimators=np.shape(X_train)[1], random_state=1),
    MLPClassifier(alpha=1),
    AdaBoostClassifier(),
    GaussianNB()]

names=["Nearest Neighbors",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes"]

multi_clfs=[]
times=[]
for clf in classifiers:
    start_time = time.time()
    multi_target_clf = MultiOutputClassifier(clf, n_jobs=-1)
    times.append(time.time() - start_time)
    multi_clfs.append(multi_target_clf.fit(X_train, Y_train))
````

````python
#transform data

dtest_1=pd.get_dummies(dtest)
Y_test = pd.DataFrame(dtest_1, columns = ['virus_H1','virus_H5','virus_H7','virus_NY','virus_Vi','virus_X31'])
X_test=pd.DataFrame(dtest_1, columns = ['P1','P3','P4','P5','P6','P7','P8','P10','P13'])
perform_table_test= pd.DataFrame(columns=('Name', 'unweighted accuracy','true positive','false positive','true negative','false negative','time for building classifier'))
perform_table_train = perform_table_test
# performance function

def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    y_actual=np.array(y_actual).flatten()
    y_hat=y_hat.flatten()  
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
    for i in range(len(y_hat)): 
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==0:
           TN += 1
    for i in range(len(y_hat)): 
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1
    return(TP, FP, TN, FN)



````

````python
# run classifier on test data and check performace of the classifier on train data

for multi_clf,name,i,perform in zip(multi_clfs,names,range(len(names)),times):
    TP, FP, TN, FN= perf_measure(Y_train, multi_clf.predict(X_train))
    
    perform_table_train.loc[i] = name,100*multi_clf.score(X_train,Y_train),100*float(TP)/(TP+FP),100*float(FP)/(TP+FP),100*float(TN)/(FN+TN),100*float(FN)/(FN+TN), perform

print perform_table_train
````

````python
#performance graph on training data
perform_table_train.drop('time for building classifier',1).plot.bar("Name")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance Graph on training data')
plt.ylabel('in percent')
plt.show()
````

````python
# run classifier on test data and check performace of the classifier on test data

for multi_clf,name,i,perform in zip(multi_clfs,names,range(len(names)),times):
    TP, FP, TN, FN= perf_measure(Y_test, multi_clf.predict(X_test))
    perform_table_test.loc[i] = name,100*multi_clf.score(X_train,Y_train),100*float(TP)/(TP+FP),100*float(FP)/(TP+FP),100*float(TN)/(FN+TN),100*float(FN)/(FN+TN), perform

print perform_table_test
````

````python
#performance graph on test data
perform_table_test.drop('time for building classifier',1).plot.bar("Name")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance Graph of test data')
plt.ylabel('in percent')
plt.show()
````

````python
#data visualization idea
#decompose into principle components and reduce dimension to 2. 
pca = decomposition.PCA(n_components=2)
pca.fit(X_train)
X_train_trans = pca.transform(X_train)
X_test_trans = pca.transform(X_test)

a=np.arange(Y_train.shape[1])+1
colorindex_train=(a*Y_train).T.sum()
colorindex_test=(a*Y_test).T.sum()
plt.xlabel("princple component 1")
plt.ylabel("princple component 2")

plt.scatter(X_train_trans[:,0],X_train_trans[:,1],c=colorindex_train/6, label="train")
plt.scatter(X_test_trans[:,0],X_test_trans[:,1],c=colorindex_test/6, marker="x", label="test")
plt.legend(loc=2)
````

````python
# vizualization idea of classifier with reduced dimensions to 2.
# build grid
numPoints=50
x_mesh=[]
mins=X_train_trans.min(axis=0)
maxs=X_train_trans.max(axis=0)
x = np.arange(mins[0], maxs[0], abs(-mins[0]+maxs[0])/50)
y = np.arange(mins[1], maxs[1],abs(-mins[1]+maxs[1])/50)
n=0
grid=[]
for i in x:
    m=0
    for j in y:
        grid.append([i,j])
        m=m+1
    n=n+1
grid=np.array(grid)
grid_test=pca.inverse_transform(grid)

#classify grid points 

multi_clfs=[]
Y_mesh_s=[]
for clf in classifiers:
    multi_target_clf = MultiOutputClassifier(clf, n_jobs=-1)
    Y_grid=multi_target_clf.fit(X_train_trans, Y_train).predict(grid)
    a=np.arange(Y_grid.shape[1])+1
    colorindex_grid=(a*Y_grid).T.sum(axis=0)
    Y_mesh_s.append(colorindex_grid)
````

````python
#plot grid data, training data and test data
for i,name  in zip(Y_mesh_s,names):
    plt.scatter(grid[:,0],grid[:,1],c=i/6,alpha=0.2)
    plt.scatter(X_train_trans[:,0],X_train_trans[:,1],c=colorindex_train/6,label="train")
    plt.scatter(X_test_trans[:,0],X_test_trans[:,1],c=colorindex_test/6, marker="x",label="test")
    plt.xlabel("princple component 1")
    plt.ylabel("princple component 2")
    plt.title(name)
    plt.legend(loc=2)
    plt.show()
````
